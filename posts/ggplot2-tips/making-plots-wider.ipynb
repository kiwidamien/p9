# Problem

We want to control the size of plots in plotnine.

# Solution

Using the `figure_size` option in `theme`. Other plotnine layout issues are covered well in [Hassan Kibirige's blog post](https://posit.co/blog/improving-the-layout-of-plotnine-graphics/).

## Example

Let's start with a standard, default plot.

```python
import plotnine as p9
from plotnine.data import mtcars
```

```python
(
    p9.ggplot(mtcars, mapping=p9.aes(x="wt", y="mpg", color="factor(cyl)"))
    + p9.geom_point()
)
```

Now we can change the title of the legend (to be something a little friendlier) and the figure size

```python
(
    p9.ggplot(mtcars, mapping=p9.aes(x="wt", y="mpg", color="factor(cyl)"))
    + p9.geom_point()
    + p9.scale_color_discrete(name="# Cylinders")
    + p9.theme(
        figure_size=(8, 4)
    )
)
```

Themes can also place legends inside the box (which can free more space)

```python
(
    p9.ggplot(mtcars, mapping=p9.aes(x="wt", y="mpg", color="factor(cyl)"))
    + p9.geom_point()
    + p9.scale_color_discrete(name="# Cylinders")
    + p9.theme(
        figure_size=(8, 4),
        legend_position=(1,1)
    )
)
```

When using a theme, place it _after_ any custom theme you want to use

```python
(
    p9.ggplot(mtcars, mapping=p9.aes(x="wt", y="mpg", color="factor(cyl)"))
    + p9.geom_point()
    + p9.scale_color_discrete(name="# Cylinders")
    + p9.theme_538()
    + p9.theme(
        figure_size=(8, 4),
        legend_position=(1,1)
    )
)
```
